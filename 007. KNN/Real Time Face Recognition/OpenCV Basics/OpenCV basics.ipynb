```python
import cv2
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

# Reading an image

```python
img = cv2.imread('sample.jpg')
```

# displaying an image - RGB Image

```python
plt.imshow(img)
plt.show()
```

# Convert from BGR to RGB

```python
n_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(n_img)
plt.title("BGR TO RGB")
plt.show()
```
