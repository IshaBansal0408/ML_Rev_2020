# Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
```

```python
y=(x-5)**2
```

# Goal

For a function f(x) we are going to find value of x that minimizes f(x)

```python
#Visualization
plt.style.use("seaborn")
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
c = 0
learn_rate = 0.1
error=[]
#take 50 steps downhill
for i in range(50):
    gradient = 2*(c-5) #differentiate y wrt x
    c = c - (learn_rate*gradient)
    cur_y=(c-5)**2
    error.append(cur_y)
    plt.scatter(c,cur_y)
```
