# Multi-variant Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# No correlation

```python
# Case : Bivariate 

# 1. Mean along x and y directions
mean1 = np.array([0.0,0.0])

# 2. Correation Matrix
corr1 = np.array([[1,0],[0,1]])

# 3. Generate Multivariant ND
dist1 = np.random.multivariate_normal(mean1,corr1,500)
```

```python
print(dist1)
```

```python
# 4. Separate x and y values
x1= dist1[:,0]
y1= dist1[:,1]

#5. Plotting x and y
plt.scatter(x1,y1)
plt.title("No correlation")
plt.show()
```

# Positive Correlation

```python
mean2 = np.array([13,15])
corr2= np.array([[3,2],[2,3]])
dist2 = np.random.multivariate_normal(mean2,corr2,500)
x2= dist2[:,0]
y2= dist2[:,1]
plt.scatter(x2,y2,color="green")
plt.title("Positive Correlation")
plt.show()
```

# Negative Correlation

```python
mean3 = np.array([5,6])
corr3 = np.array([[3,-2],[-2,3]])
dist3 = np.random.multivariate_normal(mean3,corr3,500)
x3= dist3[:,0]
y3= dist3[:,1]
plt.scatter(x3,y3,color="red")
plt.title("Negative Correlation")
plt.show()
```

# All Three in one

```python
plt.scatter(x3,y3,color="red")
plt.scatter(x2,y2,color="green")
plt.scatter(x1,y1)
plt.show()
```
