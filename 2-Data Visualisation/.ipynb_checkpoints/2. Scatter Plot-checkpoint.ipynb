# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

function 1: y=mx+c

function 2: y=x^2

```python
#Creating 2 arrays for x and y values

x=np.arange(10)
y1=x**2
y2=2*x+5
print ('x1 = ',x)
print ('y1 = ',y1)
print ('x2 = ',x)
print ('y2 = ',y2)
```

```python
plt.scatter(x,y1)
plt.show()
```

```python
plt.scatter(x,y2)
plt.show()
```

```python
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

# Styling Scatter Plots

```python
plt.style.use('seaborn')
```

```python
plt.scatter(x,y1,color='purple',label='y=x**2',marker='^')
plt.scatter(x,y2,color='green',label='y=2x+5',marker='*')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Basic Scatter Graphs')
plt.legend()
plt.show()
```

# Adjusting size of graph

```python
plt.figure(figsize=(2,2))

plt.scatter(x,y1,color='purple',label='y=x**2',marker='^')
plt.scatter(x,y2,color='green',label='y=2x+5',marker='*')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Basic Scatter Graphs')
plt.legend()

plt.show()
```
