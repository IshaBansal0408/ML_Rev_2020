# Bar Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.bar([1,2,3],[15,20,10])
plt.show()
```

```python
#Comparing two Graphs
plt.bar([1,2,3],[15,20,10])
plt.bar([1,2,3],[20,10,12])
plt.show()
```

```python
#Avoid Overlapping : Changing x value of one of the graph (displacing graph)
x=np.array([1,2,3])
plt.bar(x,[15,20,10])
plt.bar(x+0.5,[20,10,12])
plt.show()
```

```python
#Still Overlapping : Set width = displacement
x=np.array([1,2,3])
plt.bar(x,[15,20,10],width=0.5)
plt.bar(x+0.5,[20,10,12],width=0.5)
plt.show()
```

```python
#But one needs gap between two x values : Multiply in the original x values
x=np.array([1,2,3])*2
plt.bar(x,[15,20,10],width=0.5)
plt.bar(x+0.5,[20,10,12],width=0.5)
plt.show()
```

```python
x=np.array([1,2,3])*2
plt.bar(x,[15,20,10],width=0.5,label='first',color='purple')
plt.bar(x+0.5,[20,10,12],width=0.5,label='second',color='red')
plt.legend()
plt.show()
```

```python
#Changing the x value donation

x=np.array([1,2,3])*2
plt.bar(x,[15,20,10],width=0.5,label='first',color='purple',tick_label=['A','B','C'])
plt.bar(x+0.5,[20,10,12],width=0.5,label='second',color='red')
plt.legend()
plt.title("BAR GRAPHS")
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
#Changing the x value donation

x=np.array([1,2,3])*2
plt.bar(x,[15,20,10],width=0.5,label='first',color='lightgreen',tick_label=['A','B','C'])
plt.bar(x+0.5,[20,10,12],width=0.5,label='second',color='yellow')
plt.legend()
plt.title("BAR GRAPHS")
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
#Changing the x value donation

x=np.array([1,2,3])*2
plt.bar(x,[15,20,10],width=0.5,label='first',color='lightgreen',tick_label=['A','B','C'])
plt.bar(x+0.5,[20,10,12],width=0.5,label='second',color='yellow')
plt.legend()
plt.title("BAR GRAPHS")
plt.xlabel('x values')
plt.ylabel('y values')
plt.ylim(0,30)
plt.xlim(1,7)
plt.show()
```
