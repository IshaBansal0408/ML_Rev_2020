# Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
weightage=[25,20,30,25]
plt.pie(weightage)
plt.show()
```

```python
subjects=['Maths','Biology','Physics','Chemistry']
weightage=[25,20,30,25]
plt.pie(weightage,labels=subjects)
plt.show()
```

```python
subjects=['Maths','Biology','Physics','Chemistry']
weightage=[25,20,30,25]
plt.pie(weightage,labels=subjects,explode=(0,1,0,0))
#second pie is taken out by 1 distance
plt.show()
```

```python
plt.pie(weightage,labels=subjects,explode=(0,0,2,0))
#third pie is taken out by 2 distance
plt.show()
```

```python
subjects=['Maths','Biology','Physics','Chemistry']
weightage=[50,20,30,20]
plt.pie(weightage,labels=subjects,shadow=True)
plt.show()
```
