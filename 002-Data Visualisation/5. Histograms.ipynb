# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Normal Distribution

```python
xsn= np.random.randn(500) #random standard normal distribution points
#print(xsn) 
# most of the values between -1 and +1
```

Formula : x= xsn*var+mean

```python
var=5
mean=60
x=xsn*var+mean
```

```python
#print(x) 
# Most of the values will be between 55 and 65
```

```python
x=np.round(x)
```

```python
print(x)
```

# Drawing Histogram

```python
plt.style.use('seaborn')
plt.hist(x)
plt.xlabel('x range values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

# Having Two Distributions

Consider marks of two subjects : Maths and Science

```python
Xsn1=np.random.randn(100)
Xsn2=np.random.randn(100)
```

```python
var1=5
mean1=60
var2=5
mean2=40
```

```python
x1=Xsn1*var1+mean1
x2=Xsn2*var2+mean2
```

```python
plt.hist(x1)
plt.xlabel('Marks')
plt.ylabel('No of Students')
plt.title('Maths')
plt.show()
```

```python
plt.hist(x2)
plt.xlabel('Marks')
plt.ylabel('No of Students')
plt.title('Science')
plt.show()
```

```python
plt.hist(x1)
plt.hist(x2)
plt.xlabel('Marks')
plt.ylabel('No of Students')
plt.show()
```

```python
#Adding Opacity to the graphs - Overlapping is visible
# alpha - overlapping component
plt.hist(x1,alpha=0.8,label="Maths")
plt.hist(x2,alpha=0.8,label="Science")
plt.xlabel('Marks')
plt.ylabel('No of Students')
plt.title("Marks Vs No of students")
plt.legend()
plt.show()
```
