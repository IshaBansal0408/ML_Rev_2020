```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(5)
print(x)
```

```python
x_2=x**2
x_3=x**3
```

# Line plot

```python
plt.plot(x,x_2,color="Black",label="x_2")
plt.plot(x,x_3,color="Orange",label="x_3")
plt.xlabel("x")
plt.ylabel("result")
plt.title("$X^2$ vs $X^3$")
plt.legend()
plt.show()
```

# Scatter Plot

```python
plt.scatter(x,x_2,color="Green",label="x_2")
plt.scatter(x,x_3,color="Red",label="x_3")
plt.xlabel("x")
plt.ylabel("result")
plt.title("$X^2$ vs $X^3$")
plt.legend()
plt.show()
```

```python
x_1=np.random.randint(2,40,10)
y_1=np.arange(10)
```

```python
x_2=np.random.randint(2,40,10)
y_2=np.arange(10)
```

```python
plt.bar(y_1,x_1,color="Red",width=0.3)
plt.bar(y_2,x_2,color="BLUE",width=0.3)
plt.show()
```

```python
#AVOID OVERLAPPING
plt.bar(y_1,x_1,color="Red",width=0.3)
plt.bar(y_2+0.3,x_2,color="BLUE",width=0.3)
plt.show()
```

# Histograph

```python
#How to plot a histogram
#Lets first make a data which belongs to normal distribution.Normal Distribution is made for continuous random variables

#Explain random variable,Discrete random variable and continous random variable
# The first parameter means mean ,second parameter means std,third parameter means total number of values where all the values
# will belong to the normal distribution
data=np.random.normal(1,10,30000)
print(data)
print()
print(data.shape)
print()
plt.hist(data)
plt.show()
```

```python
plt.hist(data,bins=5,density="True")
plt.show()
```

# Pie chart

```python
labels=["AI","DL","DS","ML"]
counts=[40,80,60,320]
plt.pie(counts,labels=labels,shadow="True")
plt.show()
```

```python
plt.pie(counts,labels=labels,shadow="True",explode=(0,1,0,1))
plt.show()
```

```python
plt.pie(counts,labels=labels,colors=['pink','purple','blue','black'], shadow="True",explode=(0,1,1,0))
plt.show()
```
