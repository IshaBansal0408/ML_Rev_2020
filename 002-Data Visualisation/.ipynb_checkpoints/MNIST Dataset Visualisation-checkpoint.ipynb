Show a sample Image

```python
def draw_image(X,Y,i):
    plt.imshow(X[i].reshape(28,28))
    plt.title("Label:"+str(Y[i]))
    plt.show()
```

# Splitting Data for training and testing

Define Split size : 80-20 rule

```python
split= int(0.8*values.shape[0])
print("Training data size:"+str(split))
print("Testing data size:"+str(values.shape[0]-split))
```

Training Data

```python
x_train = values[:split,:]
y_train = labels[:split]
```

```python
x_train.shape
```

```python
y_train.shape
```

Testing Data

```python
x_test = values[split:,:]
y_test = labels[split:]
```

```python
y_test.shape
```

# Plotting Grid of 25 images

Concept of sub-plots
The grid we want will be of size 5x5

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x_train[i].reshape(28,28))
    plt.axis("off")
    plt.title("Label:"+str(y_train[i]))
```
