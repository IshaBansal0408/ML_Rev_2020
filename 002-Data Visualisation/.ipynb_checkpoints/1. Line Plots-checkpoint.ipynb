# Line Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

function 1: y=mx+c

function 2: y=x^2

```python
#Creating 2 arrays for x and y values

x=np.arange(10)
y1=x**2
y2=2*x+5
print ('x1 = ',x)
print ('y1 = ',y1)
print ('x2 = ',x)
print ('y2 = ',y2)
```

```python
#Plotting Line graph
plt.plot(x,y1)
plt.show() 
```

```python
plt.plot(x,y2)
plt.show()
```

# Styling Line Plots

Themes available

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
#Plotting Line graph
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#Plotting Line graph
plt.plot(x,y1,color='purple')
plt.plot(x,y2,color='green')
plt.show()
```

```python
plt.plot(x,y1,color='purple',linestyle='dotted')
plt.plot(x,y2,color='green',linestyle='dashed')
plt.show()
```

```python
plt.plot(x,y1,color='purple',marker='o')
plt.plot(x,y2,color='green',marker='*')
plt.show()
```

```python
plt.plot(x, y1, 'bo')
plt.show()
```

# Labeling in Graphs

```python
#Plotting Line graph
plt.plot(x,y1,color='purple',label='y=x**2')
plt.plot(x,y2,color='green',label='y=2x+5')
plt.legend()
plt.show()
```

```python
plt.plot(x,y1,color='purple',label='y=x**2')
plt.plot(x,y2,color='green',label='y=2x+5')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Basic Line Graphs')
plt.legend()
plt.show()
```
