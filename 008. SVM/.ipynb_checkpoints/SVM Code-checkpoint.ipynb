# SVM Implementation

# Step 1 : Data Generation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
 from sklearn.datasets import make_classification
```

```python
#Generate a random n-class classification problem
x,y=make_classification(n_classes=2,n_samples=400,
                        n_features=2,n_informative=2,n_redundant=0,
                       random_state=3,n_clusters_per_class=1)
```

```python
print(x.shape,y.shape)
```

```python
print(y)
```

```python
# The problem is that labels should be -1 and +1 not 0 and 1 cuz in case of 0 data will be lost in calculations
y[y==0]=-1
print(y)
```

```python
# Plotting data
plt.scatter(x[:,0],x[:,1],c=y,cmap="flag")
plt.show()
```

```python
class SVM:
    def __init__(self,c=1.0):
        self.c=c
        self.w=0
        self.b=0
    def Hinge_Loss_function(self,w,b,x,y):
        loss=0.0
        v1= 0.5*np.dot(w,w.T)
        loss+=v1
        
        v2=0.0
        m=x.shape[0]
        for i in range(m):
            ti=y[i]*(np.dot(w,x[i].T)+b)
            v2+=self.c*(max(0,(1-ti)))
        loss+=v2
        return loss[0][0] # Initial loss comes in form of a matrix
    def fit(self,x,y,batch_size=100,learning_rate=0.001,n_itr=300):
        n_f=x.shape[1]
        n_s=x.shape[0]
        lr=learning_rate
        c=self.c
        # initialize w and b before starting to update the values to get accurate hyperplane
        w=np.zeros((1,n_f))
        b=0
        #initial Loss
        #print(self.Hinge_Loss_function(w,b,x,y))
        
        # Training data and using final update rules for w and b
        all_loss=[]
        for i in range(n_itr):
            #current iteration loss
            l=self.Hinge_Loss_function(w,b,x,y)
            all_loss.append(l)
            # Batch Gradient Descent (given in paper) with random shuffling
            """
            Basically taking random indexes (equal to total samples) 
                -Shuffling them 
                -Distributing them into batches
                -Update W and b for the particular batch and so on.
            """
            random_ind=np.arange(n_s)
            np.random.shuffle(random_ind)
            for start in range(0,n_s,batch_size):
                #assume gradient as 0 for the batch
                grad_w=0.0
                grad_b=0.0
                #Iterate over the whole batch
                for j in range(start,start+batch_size):
                    # to make sure that we don't go beyond total no of examples
                    if(j<n_s):
                        example=random_ind[j]
                        #calculate ti
                        ti=y[example]*(np.dot(w,x[example].T)+b)
                        if(ti>1):
                            #gradient remains 0 
                            grad_w+=0
                            grad_b+=0
                        else:
                            grad_w+=c*y[example]*x[example]
                            grad_b+=c*y[example]
                #gradient for the whole batch is done
                #update values
                w=w-lr*w + lr*grad_w
                b=b+lr*grad_b
        self.w=w
        self.b=b
        return w,b,all_loss
```

```python
S=SVM()
final_w,final_b,losses=S.fit(x,y,n_itr=100)
```

```python
print("Final w: ",final_w)
print("Final b: ",final_b)

print("Initial Loss: ",losses[0])
print("Final Loss: ",losses[-1])
```

```python
plt.plot(losses)
plt.show()
```

```python
def plot_hyperplane(w1,w2,b):
    # Equation of line : w1.x1 + w2.x2 + b = 0
    # create a dummy data for x1 and calculate x2
    plt.figure(figsize=(12,12))
    
    dummy_x1 = np.linspace(-2,4,10)
    x_2 = -(w1*dummy_x1+b)/w2
    
    x_positive = -(w1*dummy_x1+b+1)/w2
    x_negative = -(w1*dummy_x1+b-1)/w2
    
    plt.scatter(x[:,0],x[:,1],c=y,cmap="flag")
    plt.plot(dummy_x1,x_2,color="Purple",label="Hyperplane WX+B=0")
    plt.plot(dummy_x1,x_positive,color="Green",label="Positive WX+B=1")
    plt.plot(dummy_x1,x_negative,color="Red",label="Negative WX+B=-1")
    plt.legend()
    plt.show()
```

```python
plot_hyperplane(final_w[0,0],final_w[0,1],final_b)
```
