# Normalise Data - 0 mean and 1 var

```python
# calculate mean of each feature 
mean = np.mean(x,axis=0)
```

```python
mean
```

```python
# calculate mean of each feature 
std = np.std(x,axis=0)
```

```python
std
```

```python
#Normalise data
new_x = (x-mean)/std
```

```python
df=pd.DataFrame(new_x)
```
